```python
from pyspark.sql.functions import *

df_silver = (
    spark.table("demo_catalog.bronze.patients_raw")
    .filter(col("patient_id").isNotNull())
    .withColumn("gender", upper(col("gender")))
    .withColumn("load_date", current_date())
)

df_silver.write.mode("overwrite").format("delta") \
    .saveAsTable("demo_catalog.silver.patients_clean")
```
