```python
df_gold = (
    spark.table("demo_catalog.silver.patients_clean")
    .groupBy("city", "disease")
    .agg(
        count("patient_id").alias("total_patients"),
        sum("bill_amount").alias("total_revenue")
    )
)

df_gold.write.mode("overwrite").format("delta") \
    .saveAsTable("demo_catalog.gold.revenue_summary")
```
