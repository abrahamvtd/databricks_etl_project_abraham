```python
from pyspark.sql.types import *

schema = StructType([
    StructField("patient_id", IntegerType()),
    StructField("name", StringType()),
    StructField("age", IntegerType()),
    StructField("gender", StringType()),
    StructField("city", StringType()),
    StructField("admission_date", DateType()),
    StructField("disease", StringType()),
    StructField("bill_amount", IntegerType())
])

df_bronze = (
    spark.read
    .schema(schema)
    .option("header", True)
    .csv("/Volumes/demo_catalog/healthcare/raw_volume/patient_50k_records.csv")
)

df_bronze.write.mode("overwrite").format("delta") \
    .saveAsTable("demo_catalog.bronze.patients_raw")
```
